```python
import pandas as pd

dataset = pd.read_csv('https://raw.githubusercontent.com/rodrigogoulart/IAC/refs/heads/main/imports-85.data.csv');
dataset
```

```python
import matplotlib.pyplot as plt
dataset.boxplot()
plt.xticks(rotation=90) # Rotate x-axis labels by 90 degrees
plt.show()
```

```python
import matplotlib.pyplot as plt

dataset.boxplot(column=['city-mpg'])
plt.title('Boxplot de City MPG')
plt.ylabel('City MPG')
plt.show()
```

```python
one_hot_dataset = pd.get_dummies(dataset,drop_first=True)
one_hot_dataset
```

```python
X = one_hot_dataset[one_hot_dataset.columns[1:]]
X
```

```python
y = one_hot_dataset['symboling']
y
```

```python
# Importando biblioteca de separação de treino / teste
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier # Importando classificador
# Cria o objeto de um classificador de Árvores de Decisão
clf = DecisionTreeClassifier(criterion='entropy',min_samples_leaf=1)
# Executa o treinamento (fit)
clf = clf.fit(X_train,y_train)
```

```python
from sklearn import tree
tree.plot_tree(clf,rounded=True,feature_names=X.columns,fontsize=3,class_names=['-3','-2','-1','0','1','2','3'],proportion=True, impurity=False, filled=True, label='all')
```

```python
# Faz a predição dos valores para teste
y_pred = clf.predict(X_test)
X_test,y_pred
```

```python
from sklearn import metrics #Importando módulo de métricas para avaliação
# Acurácia do modelo, o quanto o classificador está correto?
print("Acurácia:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, classification_report

print(classification_report(y_test, y_pred, target_names=['-2','-1','0','1','2','3']))
```

```python
ConfusionMatrixDisplay.from_estimator(
    clf, X_test, y_test, display_labels=['-2','-1','0','1','2','3'], xticks_rotation="vertical"
)
plt.tight_layout()
plt.show()
```
